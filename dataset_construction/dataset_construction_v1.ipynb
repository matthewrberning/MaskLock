```python
path = "C:/Users/OI/Desktop/MaskLockData/face_mask_detection_archive/annotations"
img_dir_path = "C:/Users/OI/Desktop/MaskLockData/face_mask_detection_archive/images"

save_path = "C:/Users/OI/Desktop/MaskLockData/"

xml_annotation_files = os.listdir(path)

xml_annotation_files = xml_annotation_files[:2]

for file in xml_annotation_files:
    #tree = ET.parse(os.path.join(path, xml_annotation_files[0]))
    tree = ET.parse(os.path.join(path, file))
    root = tree.getroot()
    
    filename = '' #collect the filename for the current image
    
    for elm in root.iter('filename'):
        #print(ET.tostring(elm).decode('utf8'))
        filename = elm.text
        
    face_num = 0 #create iterable to track faces coming out of cropping from a single image
    
    for elm in root.iter('object'): #find all 'objects' which are faces in the image
        for child in elm.findall('name'): #now find the labels for each of those faces
            #print(ET.tostring(child).decode('utf8'))
            #print(child.text)

            if child.text == 'with_mask': #if the label indicates a face with a mask work on it
                #print(ET.tostring(elm).decode('utf8'))

                #placeholders and collection of bounding box coords
                xmin = '' 
                ymin = ''
                xmax = ''
                ymax = ''

                for child in elm.findall('bndbox'):
                    #print(ET.tostring(child).decode('utf8'))

                    for i in child.findall('xmin'):
                        #print(ET.tostring(i).decode('utf8'))
                        xmin = i.text
                    for i in child.findall('ymin'):
                        #print(ET.tostring(i).decode('utf8'))
                        ymin = i.text
                    for i in child.findall('xmax'):
                        #print(ET.tostring(i).decode('utf8'))
                        xmax = i.text
                    for i in child.findall('ymax'):
                        #print(ET.tostring(i).decode('utf8'))
                        ymax = i.text

                print(xmin, ymin, xmax, ymax)
                print(filename)
                
                print(face_num)
                print("++++")
                
                #open with PIL and crop face based upon coords, with padding
                
                padding = 20
                
                im = Image.open(os.path.join(img_dir_path, filename)).convert("RGB")
                
                left = max(int(xmin)-padding, 0)
                upper = max(int(ymin)-padding, 0)
                right = min(int(xmax)+padding, im.size[0])
                lower = min(int(ymax)+padding, im.size[1])
                
                box = (left, upper, right, lower)
                region = im.crop(box)
                
                #save crop                
                region.save(os.path.join(save_path, "test_face_crop.png"), "PNG")
                
                face_num += 1 #iterate face number

                
    print("/////end of this set of faces///")
    
```
