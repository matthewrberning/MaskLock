### This next set of commands is pulling out a large number of the 'no_mask' class so that the classes will be a bit more ballanced

```python
source_path_no_mask_train = "E:/One Drive/OneDrive/Mainproject/MaskLockData/dataset/train/no_mask"
source_path_no_mask_val = "E:/One Drive/OneDrive/Mainproject/MaskLockData/dataset/val/no_mask"
```

```python
#list the files 
file_list_train = os.listdir(source_path_no_mask_train)
file_list_val = os.listdir(source_path_no_mask_val)
print("Number of images in train: ",len(file_list_train))
# print(file_list[:6])
print("Number of images in val: ",len(file_list_val))
```

```python
#shuffle the list
random.shuffle(file_list_train)
random.shuffle(file_list_val)
```

```python
#select the first 15,000 in the train set to keep
to_move_train_list = file_list_train[15000:]

print("number of train files to move: ", len(to_move_train_list))
```

```python
#select the firts 6,000 from the val set to keep
to_move_val_list = file_list_val[6000:]

print("number of train files to move: ", len(to_move_val_list))
```

```python
dest_path_no_mask_val = "E:/One Drive/OneDrive/Mainproject/MaskLockData/dataset/val/excess_uncovered_faces"
dest_path_no_mask_train = "E:/One Drive/OneDrive/Mainproject/MaskLockData/dataset/train/excess_uncovered_faces"
```

```python
for file in to_move_train_list:
    print("moving train file: ", file)
    shutil.move(os.path.join(source_path_no_mask_train, file), os.path.join(dest_path_no_mask_train, file))
```

```python
for file in to_move_val_list:
    print("moving val file: ", file)
    shutil.move(os.path.join(source_path_no_mask_val, file), os.path.join(dest_path_no_mask_val, file))
```
